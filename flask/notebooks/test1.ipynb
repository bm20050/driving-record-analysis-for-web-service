```python
import pandas as pd
```

```python
df = pd.read_csv('../data/0524_to_db.csv')
```

```python
df = df.query('일 == 1 and 차량번호 == "부산70자1894"')
df
```

```python
df['가속도max'] = df['차량속도'].apply(lambda x: -3.007e-10 * x ** 5 + 1.08775e-7 * x ** 4 + -1.46299e-5 * x ** 3 + 0.001332283 * x ** 2 -0.126973836 * x + 9.212774118)
df
```

```python
df['가속도min'] = df['차량속도'].apply(lambda x: 1.0479e-10 * x ** 5 + -4.93699e-08 * x ** 4 + 7.45908e-06 * x ** 3 + -0.000180393 * x ** 2 -0.05416834 * x + 6.010221586)
df
```

```python

velocity = [i for i in range(60)]
```

```python
temp = pd.DataFrame({'속도': velocity})
```

```python
temp['ymin'] = temp['속도'].apply(lambda x: 1.0479e-10 * x ** 5 + -4.93699e-08 * x ** 4 + 7.45908e-06 * x ** 3 + -0.000180393 * x ** 2 -0.05416834 * x + 6.010221586)
```

```python
temp['ymax'] = temp['속도'].apply(lambda x: -3.007e-10 * x ** 5 + 1.08775e-7 * x ** 4 + -1.46299e-5 * x ** 3 + 0.001332283 * x ** 2 -0.126973836 * x + 9.212774118)
```

```python
temp['y2max'] = temp['ymax'] * 2 - temp['ymin']
```

```python
temp
```

```python
import matplotlib.pyplot as plt
```

```python
d = df.query('가속도 >= 0')
d
```

```python
plt.plot(temp['속도'], temp['ymin'], linewidth=2)
plt.plot(temp['속도'], temp['ymax'], linewidth=2)
plt.plot(temp['속도'], temp['y2max'], linewidth=2)
plt.scatter(d['차량속도'], d['가속도'])
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/slice_data.csv')
```

```python
df['GPS_X'] = df['GPS_X'] / 1000000
df['GPS_Y'] = df['GPS_Y'] / 1000000
```

```python
g = df.query('번호판 == "부산70자1854"')
g.head()
```

```python
lst = []
for i in g.index:
    s = str(g['정보발생일시'][i])
    lst.append(s[:9] + '00000')

g['10분단위'] = lst
```

```python
g
```

```python
lst1 = []
tenm = []
year = []
month = []
day = []
hour = []
minute = []
lst2 = []
lst3 = []
lst4 = []
for i in g.index:
    s = g['정보발생일시'][i]
    s = str(s)

    lst1.append(s)
    tenm.append(g['10분단위'][i])
    year.append(s[:2])
    month.append(s[2:4])
    day.append(s[4:6])
    hour.append(s[6:8])
    minute.append(s[8:10])
    lst2.append(g['차량속도'][i])
    lst3.append(g['GPS_X'][i])
    lst4.append(g['GPS_Y'][i])

temp_df = pd.DataFrame(
    {'정보발생일시': lst1, '10분단위': tenm, '년': year, '월': month, '일': day, '시': hour, '분': minute, '차량속도': lst2,
     'GPS_X': lst3, 'GPS_Y': lst4}
)
```

```python
temp_df
```

```python
group_10 = temp_df.groupby(['일', '10분단위'])
group_10
```

```python
vel = group_10.mean()['차량속도']
vel = pd.DataFrame(vel)
vel
```

```python
merge_df = pd.merge(temp_df, vel, how='inner', on=["일", "10분단위"])
merge_df
```

```python
merge_df.to_csv('../data/ten_min_gps_data.csv')
```

```python
merge_df['혼잡도'] = pd.cut(merge_df.차량속도_y, bins=[-1, 15, 25, 37], labels=['정체', '서행', '원활'])
merge_df
```

```python
merge_df.to_csv('../data/ten_min_gps_data_label.csv')
```
